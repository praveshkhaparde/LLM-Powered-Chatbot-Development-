```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
```

```python
df.drop(columns= ['id', 'Unnamed: 32'], inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
## Numpy Arrays to PyTorch tensors

X_train_tensor = torch.from_numpy(X_train)
y_train_tensor = torch.from_numpy(y_train)
X_test_tensor = torch.from_numpy(X_test)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

## Defining the Model

```python
class MySimleNN():
    ## Single Neuron
    def __init__(self, X):
        self.W = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
        self.b = torch.zeros(1, dtype = torch.float64, requires_grad=True)
        ## requires_grad means this is to be trained
    
    def forward(self, X):
        z = X @ self.W + self.b
        y_pred = torch.sigmoid(z)
        return y_pred

    def loss_function(self, y_pred, y_train_tensor):
        ## Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

        ## Calculate loss
        loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
        return loss
```

### Hyperparameters

```python
learning_rate = 0.1
epochs = 25
```

### Training Pipeline

```python
## creating model
model = MySimleNN(X_train_tensor)
model.b
```

```python
## Training the model

## Repeat(number of epochs)
##      forward pass
##      loss calculation
##      backward pass
##      Parameter Update
for epoch in range(epochs):
    y_pred = model.forward(X_train_tensor)
    loss = model.loss_function(y_pred, y_train_tensor)
    loss.backward()
    with torch.no_grad():
        model.W -= learning_rate * model.W.grad
        model.b -= learning_rate * model.b.grad

    model.W.grad.zero_()
    model.b.grad.zero_()

    print(f'Epoch: {epoch + 1}, Loss: {loss}')
```

```python
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
```
